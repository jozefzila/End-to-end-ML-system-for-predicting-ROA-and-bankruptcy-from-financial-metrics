# Regression Modeling: Predicting Return on Assets (ROA)

This notebook builds and evaluates regression models that predict the **ROA (Return on Assets)** of companies based on financial metrics.

ROA is treated as a continuous variable.

The notebook includes:
- Data exploration and preparation 
- Training and evaluation of multiple regression models:
  - Histogram Gradient Boosting Regressor
  - XGBoost Regressor
  - Random Forest Regressor
  - Regression metrics, their evaluation and SHAP interpretation
  - Prediction of ROA on a given company

```python
# Define quantile limits (for handling outliers - we have examined that 3% from upper and lower quantile is best)
lower_quantile = 0.03
upper_quantile = 0.97

# Calculate lower and upper bounds for each column
lower_bounds = data.quantile(lower_quantile)
upper_bounds = data.quantile(upper_quantile)

# Apply row-wise filtering
# Keep rows where all values are within their respective quantile bounds
data1 = data[
    ~((data < lower_bounds) | (data > upper_bounds)).any(axis=1)
]
```

### **Model Training and Evaluation** 

```python


# Split into features and target
X = data1.drop(["ROA", "ROI"], axis=1) #We exclude ROI as it is perfectly correlated with dependent variable
y = data1["ROA"]
n, p = X.shape  # Number of samples and features

# Stratified K-Fold Binning
y_binned = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')\
    .fit_transform(y.to_numpy().reshape(-1, 1)).ravel()
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Initialize models
models = {
    "HistGradientBoosting": HistGradientBoostingRegressor(max_iter=300, max_depth=12, min_samples_leaf=5, learning_rate=0.01, random_state=42),
    "RandomForest": RandomForestRegressor(n_estimators=200, max_depth=12, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=200, max_depth=12, learning_rate=0.01, random_state=42)
}

# Function to calculate adjusted R^2
def adjusted_r2(r2, n, p):
    return 1 - ((1 - r2) * (n - 1) / (n - p - 1))

# Evaluate models
results = {}
r2_scores_all, rmse_scores_all, adj_r2_scores_all = {}, {}, {}

for name, model in models.items():
    r2_scores, rmse_scores, adj_r2_scores = [], [], []
    
    for train_idx, test_idx in skf.split(X, y_binned):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        r2 = r2_score(y_test, y_pred)
        r2_scores.append(r2)
        rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))
        adj_r2_scores.append(adjusted_r2(r2, n, p))
    
    # SHAP Analysis (after cross-validation, using the last fold's model as an example)
    explainer = shap.Explainer(model, X_train)
    shap_values = explainer(X_test)
    
    # SHAP Summary Plot
    plt.figure(figsize=(10, 6))
    shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)
    plt.title(f"SHAP Feature Importance - {name}")
    plt.show()
    
    # SHAP Beeswarm Plot (detailed view)
    plt.figure(figsize=(10, 6))
    shap.summary_plot(shap_values, X_test, show=False)
    plt.title(f"SHAP Beeswarm Plot - {name}")
    plt.show()
    
    results[name] = {
        "Holdout R2": np.mean(r2_scores), 
        "Avg RMSE": np.mean(rmse_scores), 
        "Avg R2": np.mean(r2_scores),
        "Avg Adjusted R2": np.mean(adj_r2_scores)
    }
    r2_scores_all[name] = r2_scores
    rmse_scores_all[name] = rmse_scores
    adj_r2_scores_all[name] = adj_r2_scores

# Convert results to DataFrame
results_df = pd.DataFrame(results).T
print(results_df)

# Plot R2 comparison
plt.figure(figsize=(8, 5))
for name, scores in r2_scores_all.items():
    plt.plot(range(1, 11), scores, label=name)
plt.xlabel("Fold")
plt.ylabel("R^2 Score")
plt.title("R^2 Models Comparison")
plt.legend()
plt.show()

# Plot Adjusted R2 comparison
plt.figure(figsize=(8, 5))
for name, scores in adj_r2_scores_all.items():
    plt.plot(range(1, 11), scores, label=name)
plt.xlabel("Fold")
plt.ylabel("Adjusted R^2 Score")
plt.title("Adjusted R^2 Models Comparison")
plt.legend()
plt.show()

# Plot RMSE comparison
plt.figure(figsize=(8, 5))
for name, scores in rmse_scores_all.items():
    plt.plot(range(1, 11), scores, label=name)
plt.xlabel("Fold")
plt.ylabel("RMSE")
plt.title("RMSE Models Comparison")
plt.legend()
plt.show()
```

**Compare Actual vs Predicted values**

```python
# Store predictions across all folds
all_predictions = {name: {'y_valid': [], 'y_pred': []} for name in models.keys()}

# Modify evaluation loop to collect predictions
for name, model in models.items():
    r2_scores, rmse_scores, adj_r2_scores = [], [], []
    
    for train_idx, test_idx in skf.split(X, y_binned):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        r2 = r2_score(y_test, y_pred)
        r2_scores.append(r2)
        rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))
        adj_r2_scores.append(adjusted_r2(r2, n, p))
        
        # Collect actual and predicted values
        all_predictions[name]['y_valid'].extend(y_test.values)
        all_predictions[name]['y_pred'].extend(y_pred)
    
    results[name] = {
        "Holdout R2": np.mean(r2_scores), 
        "Avg RMSE": np.mean(rmse_scores), 
        "Avg R2": np.mean(r2_scores),
        "Avg Adjusted R2": np.mean(adj_r2_scores)
    }
    r2_scores_all[name] = r2_scores
    rmse_scores_all[name] = rmse_scores
    adj_r2_scores_all[name] = adj_r2_scores

# Plot Actual vs. Predicted ROA for each model using all folds
for name, data in all_predictions.items():
    y_valid = np.array(data['y_valid'])
    y_pred = np.array(data['y_pred'])
    
    # Sort the actual and predicted values
    y_valid_sorted = np.sort(y_valid)
    sorted_indices = np.argsort(y_valid)
    y_pred_sorted = y_pred[sorted_indices]
    
    # Compute absolute prediction error
    error = np.abs(y_valid_sorted - y_pred_sorted)
    
    # Select the 10 worst predictions
    worst_indices = np.argsort(-error)[:10]
    
    plt.figure(figsize=(12, 6))
    plt.plot(y_valid_sorted, label="Actual ROA", linestyle="-", marker="o", color="blue")
    plt.plot(y_pred_sorted, label="Predicted ROA", linestyle="--", marker="s", color="khaki")
    
    # Highlight worst predictions in red
    plt.scatter(worst_indices, y_pred_sorted[worst_indices], color="red", label="10 Worst Predictions", zorder=3)
    
    # Labels & Grid
    plt.xlabel("Number of ROA values")
    plt.ylabel("ROA")
    plt.title(f"Actual vs. Predicted ROA - {name}")
    plt.legend()
    plt.grid(True)
    plt.show()
```
